# D. Create two frequencies 5Hz and 50Hz sin signals Draw FFT components

```python
a = 5
b = 50
c = 500
d = 1
e = np.linspace(0, d, c * d, endpoint=False)
```

```python
signal_a = np.sin(2 * np.pi * a * e)
signal_b = np.sin(2 * np.pi * b * e)
```

```python
fft_vals_a = np.fft.fft(signal_a)  # FFT for a = 5
fft_freqs_a = np.fft.fftfreq(len(signal_a), 1 / c)
```

```python
fft_vals_b = np.fft.fft(signal_b)  # FFT for b = 50
fft_freqs_b = np.fft.fftfreq(len(signal_b), 1 / c)
```

```python
fft_df_a = pd.DataFrame({'Frequency': fft_freqs_a, 'Amplitude': np.abs(fft_vals_a)})
fft_df_b = pd.DataFrame({'Frequency': fft_freqs_b, 'Amplitude': np.abs(fft_vals_b)})
```

```python
fft_df_a = fft_df_a[fft_df_a['Frequency'] > 0]
fft_df_b = fft_df_b[fft_df_b['Frequency'] > 0]
```

```python
plt.figure(figsize=(10, 5))

# Plotting FFT for the a = 5 Hz signal
plt.subplot(1, 2, 1)
plt.plot(fft_df_a['Frequency'], fft_df_a['Amplitude'])
plt.title("FFT of a = 5 Hz Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plotting FFT for the b = 50 Hz signal
plt.subplot(1, 2, 2)
plt.plot(fft_df_b['Frequency'], fft_df_b['Amplitude'])
plt.title("FFT of b = 50 Hz Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# E. Plotting normal random numbers and show that addition of two Gaussian is Gaussian and addition of all is Gaussian

```python
n = 10000 
mean = 0
sd=1
```

```python
g1 = np.random.normal(mean, sd, n)
g2 = np.random.normal(mean, sd, n)
g_sum = g1 + g2
```

```python
n_gaussians = 5
sum_n_gaussians = np.zeros(n)
for _ in range(n_gaussians):
    sum_n_gaussians += np.random.normal(mean, sd, n)
```

```python
df = pd.DataFrame({
    'g1': g1,
    'g2': g2,
    'g_sum': g_sum,
    'sum_n_gaussians': sum_n_gaussians
})
```

```python
# Plotting histograms
plt.figure(figsize=(12, 8))

# Plot g1 and g2
plt.subplot(2, 2, 1)
plt.hist(g1, bins=50, alpha=0.7, label='g1')
plt.hist(g2, bins=50, alpha=0.7, label='g2')
plt.title('g1 & g2')
plt.legend()

# Plot sum of two Gaussians
plt.subplot(2, 2, 2)
plt.hist(g_sum, bins=50, alpha=0.7, color='g', label='Sum of Two Gaussians')
plt.title('Sum of Two Gaussians')
plt.legend()

# Plot sum of multiple Gaussians
plt.subplot(2, 2, 3)
plt.hist(sum_n_gaussians, bins=50, alpha=0.7, color='r', label='Sum of Multiple Gaussians')
plt.title('Sum of Multiple Gaussians')
plt.legend()

plt.tight_layout()
plt.show()
```
